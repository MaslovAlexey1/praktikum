Обозначения:

- $X$ — матрица признаков (нулевой столбец состоит из единиц)

- $y$ — вектор целевого признака

- $P$ — матрица, на которую умножаются признаки

- $w$ — вектор весов линейной регрессии (нулевой элемент равен сдвигу)

Предсказания:

$$
a = Xw
$$

Задача обучения:

$$
w = \arg\min_w MSE(Xw, y)
$$

Формула обучения:

$$
w = (X^T X)^{-1} X^T y
$$

**Ответ:** не изменится

**Обоснование:**
Посмотрим к чему мы придем умножив X на обратимую матрицу A<br>
Учитывая, что $ (ABC)^{-1} = C^{-1}B^{-1}A^{-1} $ и $ (AB)^T = B^T A^T $ <br><br>
$$
w_A = ((X A)^T (X A))^{-1} (X A)^T y = (A^T X^T X A)^{-1} A^T X^T y = A^{-1} X^{-1} (X^T)^{-1} (A^T)^{-1} A^T X^T y = ...
$$<br>
Так как матрица A обратимая, то $ (A^T)^{-1} A^T = E $ и:<br><br>
$$
... = A^{-1} X^{-1} (X^T)^{-1} E  X^T y = A^{-1} (X^T X)^{-1} X^T y = A^{-1} w
$$

Получается, что если мы умножим исходные параметры на обратимую матрицу, то параметры линейной регрессии просто умножатся на обратную матрицу

## 4. Проверка алгоритма

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

features = df.drop(['insurance'], axis=1)
target = df['insurance']

model = LinearRegression()
model.fit(features, target)
predictions = model.predict(features)

print('Предсказания модели: {}'.format(predictions))
print('R2_score {:.4f}'.format(r2_score(target, predictions)))

```

```python
from numpy.linalg import inv 

def conversion(features):
    conv_matrix = np.random.rand(features.shape[1], features.shape[1])
    features = features @ conv_matrix
    return features, conv_matrix

features = df.drop(['insurance'], axis=1)
target = df['insurance']

features, conv_matrix = conversion(features)

model = LinearRegression()
model.fit(features, target)
predictions = model.predict(features)
print('Предсказания модели после преобразования: {}'.format(predictions))
print('R2_score {:.4f}'.format(r2_score(target, predictions)))

print('Матрица преобразований:')
print(conv_matrix)


```

Убедимся что преобразование обратимое

```python
features = df.drop(['insurance'], axis=1)
print('Исходные данные')
print(features)
features, conv_matrix = conversion(features)
print('Преобразованные данные')
print(features)
print('Данные после обратного преобразования')
features = features @ inv(conv_matrix)
print(features.round())
```

R2_score в обоих случаях совпали. Выше в формулах у меня выходило что коэффициенты линейной модели будут отличаться на $A^{-1}$, но даже по размерности видно что это невозможно. Так и не нашел где я ошибаюсь в формулах :( Матричное преобразования вообще никак не повлияло на коэффициенты $ w $
